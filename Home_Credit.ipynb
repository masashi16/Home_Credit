■ dataset:

1. HomeCredit_columns_description.csv

8. POS_CASH_balance.csv
    - クライアントがHome Creditで持っていた以前のPOSの月次残高のスナップショットと，現金ローン

9. previous_application.csv
    - 過去のHome Credit社のクライアント情報
